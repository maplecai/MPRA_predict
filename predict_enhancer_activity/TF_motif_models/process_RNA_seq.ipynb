```python
RNA_df = pd.read_csv("/home/hxcai/cell_type_specific_CRE/data/RNAseq/57epigenomes_RPKM_log10.csv", sep='\t')
RNA_df
```

```python
gene_info_df = pd.read_csv('/home/hxcai/cell_type_specific_CRE/data/RNAseq/Ensembl_v65.Gencode_v10.ENSG.gene_info.txt', sep='\t', header=None,  on_bad_lines='skip')
gene_info_df.columns = ['gene_id', 'chr', 'start', 'end', 'strand', 'gene_type', 'gene_symbol', 'gene_description']
gene_info_df
```

```python
from Bio import motifs

file_path = "data/JASPAR2024_CORE_vertebrates_non-redundant_pfms_jaspar.txt"
with open(file_path) as f:
    motif_list = motifs.parse(f, "jaspar")

TF_names = [motif.name for motif in motif_list]
TF_names
```

```python
single_TF_names = [TF for TF in TF_names if TF.count(':') == 0]
print(len(single_TF_names))
```

```python
TF_info_df = gene_info_df[gene_info_df['gene_symbol'].isin(TF_names) & gene_info_df['gene_id'].isin(RNA_df['gene_id'])]
TF_info_df
```

```python
TF_RNA_df = pd.merge(TF_info_df[['gene_id', 'gene_symbol']], RNA_df, on='gene_id', how='inner')
```
