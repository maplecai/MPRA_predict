```python
split_column = 'chr'
split_dict = {
    'train': [f'chr{i}' for i in range(3, 23)],
    'valid': ['chr1'],
    'test': ['chr2'],
    'designed': ['chr0'],
}

data_split_dict = {key: main_df[main_df[split_column].isin(values)] for key, values in split_dict.items()}
```

```python
for cell_type in ['HepG2', 'K562', 'SKNSH']:
    for data_split in ['test']:
        condition = data_split_dict[data_split]
        x = TF_counts_df.sum(axis=1)[condition]
        y = main_df[condition][cell_type]
        r = pearson(x, y)
        print(cell_type, data_split, f'total r = {r:.3f}')

        condition = data_split_dict[data_split] & ~main_df['diff']
        x = TF_counts_df.sum(axis=1)[condition]
        y = main_df[condition][cell_type]
        r = pearson(x, y)
        print(cell_type, data_split, f'const r = {r:.3f}')

        condition = data_split_dict[data_split] & main_df['diff']
        x = TF_counts_df.sum(axis=1)[condition]
        y = main_df[condition][cell_type]
        r = pearson(x, y)
        print(cell_type, data_split, f'diff r = {r:.3f}')
```

```python
def remove_nan(x, y):
    if x.shape[0] == 1:
        mask = ~np.isnan(x) & ~np.isnan(y)
    else:
        mask = ~np.isnan(x).any(axis=1) & ~np.isnan(y)
    x, y = x[mask], y[mask]
    return x, y
```

```python
for cell_type in ['HepG2', 'K562']:
    print(f'model trained on {cell_type}')
    data_split = 'train'

    condition = data_split_dict[data_split]
    X_train = TF_counts_df[condition]
    y_train = main_df[condition][cell_type]

    X_train, y_train = remove_nan(X_train, y_train)
    if len(y_train) == 0:
        continue

    model = Lasso(alpha=1e-3)
    # model = MLPRegressor(hidden_layer_sizes=10, alpha=1e-3)
    model.fit(X_train, y_train)

    for cell_type in ['HepG2', 'K562', 'SKNSH']:
        for data_split in ['test', 'designed']:

            condition = data_split_dict[data_split]
            X_valid = TF_counts_df[condition]
            y_valid = main_df[condition][cell_type]

            X_valid, y_valid = remove_nan(X_valid, y_valid)
            if len(y_valid) == 0:
                continue
            y_pred = model.predict(X_valid)
            r = pearson(y_valid, y_pred)
            print(cell_type, data_split, len(y_valid), f'total r = {r:.3f}')




            condition = data_split_dict[data_split] & ~main_df['diff']
            X_valid = TF_counts_df[condition]
            y_valid = main_df[condition][cell_type]

            X_valid, y_valid = remove_nan(X_valid, y_valid)
            if len(y_valid) == 0:
                continue
            y_pred = model.predict(X_valid)
            r = pearson(y_valid, y_pred)
            print(cell_type, data_split, len(y_valid), f'const r = {r:.3f}')




            condition = data_split_dict[data_split] & main_df['diff']
            X_valid = TF_counts_df[condition]
            y_valid = main_df[condition][cell_type]
            
            X_valid, y_valid = remove_nan(X_valid, y_valid)
            if len(y_valid) == 0:
                continue
            y_pred = model.predict(X_valid)
            r = pearson(y_valid, y_pred)
            print(cell_type, data_split, len(y_valid), f'diff r = {r:.3f}')


    print()
```
