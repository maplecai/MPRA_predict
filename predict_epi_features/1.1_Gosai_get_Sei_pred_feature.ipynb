```python
MPRA_df = pd.read_csv('../data/Gosai_MPRA/Gosai_MPRA_Sei_feature_0529.csv')
MPRA_df
```

```python
# 实验说明需要标准化

data_split_dict = {}
data_split_dict['total'] = np.ones(len(MPRA_df), dtype=bool)
data_split_dict['chrom'] = MPRA_df['chr'].isin([f'chr{i}' for i in (list(range(1, 23)) + ['X', 'Y'])])
data_split_dict['designed'] = ~data_split_dict['chrom']

data_split_dict['train'] = MPRA_df['chr'].isin([f'chr{i}' for i in (list(range(3, 23)) + ['X', 'Y'])])
data_split_dict['valid'] = MPRA_df['chr'].isin(['chr1'])
data_split_dict['test'] = MPRA_df['chr'].isin(['chr2'])





# normalize between different cell types and assays
assays = ['DNase', 'H3K4me3', 'H3K27ac', 'H3K27me3', 'H3K4me1', 'H3K36me3', 'H3K9me3', 'CTCF']
cell_types = ['K562', 'HepG2', 'SK-N-SH', 'HCT116', 'A549']

norm_params = pd.DataFrame()
for i, cell_type in enumerate(cell_types):
    for j, assay in enumerate(assays):
        
        x = MPRA_df[f'{cell_type}_{assay}']
        condition = data_split_dict['train']
        x_train = x[condition]

        x_mean = np.mean(x_train)
        x_std = np.std(x_train)
        x = (x - x_mean) / x_std

        norm_params.loc['mean', f'{cell_type}_{assay}'] = x_mean
        norm_params.loc['std', f'{cell_type}_{assay}'] = x_std

        MPRA_df[f'{cell_type}_{assay}'] = x
MPRA_df
```

```python
MPRA_df = pd.read_csv('../data/Gosai_MPRA/Gosai_MPRA_Sei_feature_0529_norm.csv')

data_split_dict = {}
data_split_dict['total'] = np.ones(len(MPRA_df), dtype=bool)

data_split_dict['chrom'] = MPRA_df['chr'].isin([f'chr{i}' for i in (list(range(1, 23)) + ['X', 'Y'])])
data_split_dict['designed'] = ~data_split_dict['chrom']

data_split_dict['train'] = MPRA_df['chr'].isin([f'chr{i}' for i in (list(range(3, 23)) + ['X', 'Y'])])
data_split_dict['valid'] = MPRA_df['chr'].isin(['chr1'])
data_split_dict['test'] = MPRA_df['chr'].isin(['chr2'])

cell_types = ['K562', 'HepG2', 'SK-N-SH', 'HCT116', 'A549']
assays = ['DNase', 'H3K4me3', 'H3K27ac', 'H3K27me3', 'H3K4me1', 'H3K36me3', 'H3K9me3', 'CTCF']
```

```python
for cell_type in ['K562', 'HepG2', 'SK-N-SH']:
    condition = data_split_dict['train']

    x_train = MPRA_df.loc[condition, cell_type]
    x_mean = x_train.mean()
    x_std = x_train.std()

    # 对整个 column 做归一化（可以包含 val/test，只是用训练集参数）
    x_all = MPRA_df[cell_type]
    z_score = (x_all - x_mean) / x_std

    MPRA_df[f'{cell_type}'] = z_score
```

```python
### 归一化标签时，因为 HCT116 的序列数量和 K562 不一样多，所以只能按照 K562 子集的分布归一化
```

```python
for cell_type in ['HCT116', 'A549']:
    # 找出训练集中该 cell type 有值的样本（索引）
    condition = MPRA_df[cell_type].notna() & data_split_dict['train']

    # 拿到该 cell type 的训练值
    x_train = MPRA_df.loc[condition, cell_type]
    x_mean = x_train.mean()
    x_std = x_train.std()

    # 拿到 K562 在相同子集上的值
    k562_train = MPRA_df.loc[condition, 'K562']
    k562_mean = k562_train.mean()
    k562_std = k562_train.std()

    # 对整个 column 做归一化（可以包含 val/test，只是用训练集参数）
    x_all = MPRA_df[cell_type]
    z_score = (x_all - x_mean) / x_std
    x_adjust = z_score * k562_std + k562_mean

    MPRA_df[f'{cell_type}'] = x_adjust
```
