```python
from pyfaidx import Fasta
import pandas as pd


file_name = '../data/GTEx_eQTL/Enformer_processed/data_gtex_fine_vcf_Whole_Blood_pos.vcf'
df = pd.read_csv(file_name, sep='\t', names=['chr', 'pos', 'variant_id', 'ref', 'alt', '1', '2', '3'], comment='#')


# 加载参考基因组
fasta = Fasta('../../../genome/hg38.fa')


seqs = []
for _, row in tqdm(df.iterrows(), total=len(df)):
    # 解析变异ID: chr1_123456_G_A_b37
    variant_id = row['variant_id']
    chrom, pos, ref, alt, _ = variant_id.split('_')
    pos = int(pos)
    
    # 提取参考序列
    start = pos - 100 # pyfaidx 索引从0开始
    end = pos + 99
    
    try:
        seq = fasta[chrom][start:end].seq.upper()
        seqs.append({
            'variant_id': variant_id,
            'seq': seq,
            'chrom': chrom,
            'pos': pos,
            'ref': ref,
            'alt': alt,
            'allele': 'ref',
            'casual_label': 1,
        })
        assert seq[99] == ref

        alt_seq = seq[:99] + alt + seq[100:]
        seqs.append({
            'variant_id': variant_id,
            'seq': alt_seq,
            'chrom': chrom,
            'pos': pos,
            'ref': ref,
            'alt': alt,
            'allele': 'alt',
            'casual_label': 1,
        })


    except KeyError:
        print(f"Chromosome {chrom} not found in reference.")
        continue














file_name = '../data/GTEx_eQTL/Enformer_processed/data_gtex_fine_vcf_Whole_Blood_neg.vcf'
df = pd.read_csv(file_name, sep='\t', names=['chr', 'pos', 'variant_id', 'ref', 'alt', '1', '2', '3'], comment='#')





for _, row in tqdm(df.iterrows(), total=len(df)):
    # 解析变异ID: chr1_123456_G_A_b37
    variant_id = row['variant_id']
    chrom, pos, ref, alt, _ = variant_id.split('_')
    pos = int(pos)
    
    # 提取参考序列
    start = pos - 100 # pyfaidx 索引从0开始
    end = pos + 99
    
    try:
        seq = fasta[chrom][start:end].seq.upper()
        seqs.append({
            'variant_id': variant_id,
            'seq': seq,
            'chrom': chrom,
            'pos': pos,
            'ref': ref,
            'alt': alt,
            'allele': 'ref',
            'casual_label': 0,
        })
        assert seq[99] == ref

        alt_seq = seq[:99] + alt + seq[100:]
        seqs.append({
            'variant_id': variant_id,
            'seq': alt_seq,
            'chrom': chrom,
            'pos': pos,
            'ref': ref,
            'alt': alt,
            'allele': 'alt',
            'casual_label': 0,
        })


    except KeyError:
        print(f"Chromosome {chrom} not found in reference.")
        continue























seq_df = pd.DataFrame(seqs)
seq_df
```
