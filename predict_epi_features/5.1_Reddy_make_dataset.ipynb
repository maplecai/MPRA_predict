```python
cell_types = ['K562', 'Jurkat', 'THP-1']

pearson_df = pd.DataFrame(index=cell_types, columns=cell_types)
for cell_type in cell_types:
    for cell_type_2 in cell_types:
        x = df[f'{cell_type}']
        y = df[f'{cell_type_2}']
        r, p = pearson(x, y)
        pearson_df.loc[cell_type, cell_type_2] = r
print(pearson_df)
```
