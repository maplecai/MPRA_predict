```python
MPRA_df = pd.read_csv('../data/SirajMPRA/SirajMPRA_562654_norm.csv')
MPRA_df
```

```python
feature_matrix = np.load('Sei_Siraj_feature_matrix_norm.npy')
feature_matrix.shape
```

```python
cell_types = ['HepG2', 'K562', 'SK-N-SH', 'A549', 'HCT116']
assays = ['DNase', 'H3K4me1', 'H3K4me3', 'H3K9me3', 'H3K27me3', 'H3K27ac', 'H3K36me3', 'CTCF']


df_list = []
for i, cell_type in enumerate(cell_types):
    df = MPRA_df[['description', 'seq', 'chr', 'is_ref', 'pos', 'ref', 'alt', 'start', 'end']].copy()
    df['cell_type'] = cell_type
    df['exp'] = MPRA_df[cell_type]
    df[assays] = feature_matrix[:, i]
    df_list.append(df)

MPRA_df_concat = pd.concat(df_list, axis=0)
MPRA_df_concat
```

```python
MPRA_df_concat.to_csv('../data/SirajMPRA/SirajMPRA_562654_concat.csv', index=False)
```
