```python
import sys
sys.path.append('..')
from MPRA_predict.utils import *
```

```python
Gosai_df = pd.read_csv('../data/Gosai_MPRA/Gosai_MPRA_len200.csv')
Gosai_df
```

```python
Siraj_df = pd.read_csv('../data/SirajMPRA/SirajMPRA_562654.csv')
Siraj_df
```

```python
Siraj_df['allele'] = Siraj_df['is_ref'].apply(lambda x: 'R' if x else 'A')
Siraj_df['ID'] = Siraj_df[['description', 'allele']].astype(str).agg(':'.join, axis=1)
Siraj_df
```

```python
Siraj_df = Siraj_df[['ID', 'K562', 'HepG2', 'SK-N-SH', 'A549', 'HCT116']]
merged_df = pd.merge(Gosai_df, Siraj_df, on='ID', how='left')
merged_df
```

```python
x = merged_df['HepG2_x']
y = merged_df['HepG2_y']
print(pearson(x, y))
sns.scatterplot(x=x, y=y, s=1)

x = merged_df['K562_x']
y = merged_df['K562_y']
print(pearson(x, y))
sns.scatterplot(x=x, y=y, s=1)

x = merged_df['SK-N-SH_x']
y = merged_df['SK-N-SH_y']
print(pearson(x, y))
sns.scatterplot(x=x, y=y, s=1)
```

```python
merged_df = merged_df.rename(columns={
    'HepG2_x': 'HepG2',
    'K562_x': 'K562',
    'SK-N-SH_x': 'SK-N-SH',
})
merged_df = merged_df[['ID', 'chr', 'pos', 'ref', 'alt', 'allele', 'K562', 'HepG2', 'SK-N-SH', 'A549', 'HCT116', 'seq']]
merged_df.to_csv('../data/Gosai_MPRA/GosaiSirajMPRA.csv', index=False)
```
