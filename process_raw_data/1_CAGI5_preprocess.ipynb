```python
metadata_df[['chr', 'pos']] = metadata_df['Genomic coordinates (GRCh37)'].str.split(':', expand=True)
metadata_df[['start', 'end']] = metadata_df['pos'].str.split('-', expand=True)
metadata_df['start'] = metadata_df['start'].str.replace(',', '').astype(int)
metadata_df['end'] = metadata_df['end'].str.replace(',', '').astype(int)
metadata_df
```

```python
genome = Fasta('/home/hxcai/genome/hg19.fa')

# metadata_df['seq'] = None

for i, row in tqdm(metadata_df.iterrows(), total=len(metadata_df)):

    chr, start, end = row['chr'], row['start'], row['end']
    if row['Name'] == 'IRF6':
        end = end - 1            ######## very important!!!
    if row['Name'] == 'MYC (rs11986220)':
        end = end + 1
        
    seq = genome[chr][start-1 :end].seq.upper()     ######## very important!!!
    metadata_df.loc[i, 'seq'] = seq
metadata_df
```

```python
metadata_df['end'] - metadata_df['start'] - metadata_df['Construct size (bp)']
```

```python
metadata_dict = {row['Other_Name']: row['Name'] for i, row in metadata_df.iterrows()}
new_dict = {}
for k in metadata_dict:
    if ',' in k:
        for new_k in k.replace(' ', '').split(','):
            new_dict[new_k] = metadata_dict[k]
    else:
        new_dict[k] = metadata_dict[k]
new_dict
```

```python
# MPRA_df['seq'] = None

for i, row in tqdm(MPRA_df.iterrows(), total=len(MPRA_df)):
    element = row['Element']
    name = new_dict[element]

    meta_row = metadata_df[metadata_df['Name'] == name]
    if len(meta_row) != 1:
        print(f"Meta info not found or multiple rows for {name}")
        continue
    meta_row = meta_row.iloc[0]

    cell_type = meta_row['Cell line']
    chr, start, end, seq = meta_row[['chr', 'start', 'end', 'seq']]
    start, end = int(start), int(end)

    # print(meta_row)

    pos, ref, alt = row[['Position', 'Ref', 'Alt']]
    start, end, pos = int(start), int(end), int(pos)
    relative_pos = pos - start

    if relative_pos < 0 or relative_pos >= len(seq):
        print(relative_pos)
        continue

    if ref != seq[relative_pos]:
        print(relative_pos, ref)
        continue

    if alt == '-':
        seq = seq[:relative_pos] + seq[relative_pos+1:]
    else:
        seq = seq[:relative_pos] + alt + seq[relative_pos+1:]
    
    MPRA_df.loc[i, 'seq'] = seq
    MPRA_df.loc[i, 'cell_type'] = cell_type
```

```python
MPRA_df = MPRA_df.rename(columns={
    'Chromosome': 'chr', 
    'Position': 'pos', 
    'Ref': 'ref', 
    'Alt': 'alt', 
    'Value': 'value',
    'P-Value': 'p-value',
    'Element': 'element',
    })


MPRA_df = MPRA_df[['chr', 'pos', 'ref', 'alt', 'value', 'p-value', 'element', 'cell_type', 'seq']]
MPRA_df
```

```python
MPRA_df = MPRA_df[MPRA_df['seq'].notna()].reset_index(drop=True)

MPRA_df
```
