```python
df_large_scale = pd.read_excel('supp_materials/media-3.xlsx', header=3, sheet_name='HepG2 large-scale')
df_large_scale
```

```python
df_summary_data = pd.read_excel('supp_materials/media-4.xlsx', sheet_name='HepG2_summary_data')
df_summary_data
```

```python
df_HepG2 = pd.merge(df_large_scale, df_summary_data, on='name', how='inner')
df_HepG2
```

```python
df_HepG2 = df_HepG2.rename(columns={'230nt sequence (15nt 5\' adaptor - 200nt element - 15nt 3\' adaptor)': 'seq',
                                    'mean': 'exp',
                                    'chr.hg38': 'chr',
                                    'start.hg38':'start',
                                    'stop.hg38': 'end',
                                    'str.hg38': 'strand'})
df_HepG2 = df_HepG2[['seq', 'exp', 'chr', 'start', 'end', 'strand']]
df_HepG2
```

```python
df_HepG2.to_csv('Agarwal_HepG2.csv', index=False)
```
