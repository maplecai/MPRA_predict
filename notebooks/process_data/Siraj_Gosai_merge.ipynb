```python
Gosai_df = pd.read_csv('../../data/GosaiMPRA/supp/SupTable 2 - UKBB_GTEX_CODA_averaged_no_cutoffs.txt', delimiter='\t', low_memory=False)
total_cols = Gosai_df.columns.tolist()
reorder_cols = ['IDs', 'nt_sequence', 'chr', 'OL', 'data_project', 'class']
other_cols = [x for x in total_cols if x not in reorder_cols]
Gosai_df = Gosai_df[reorder_cols + other_cols]
Gosai_df.rename(columns={'nt_sequence': 'seq'}, inplace=True)
Gosai_df
```

```python
# Siraj_df = pd.read_csv("../../data/SirajMPRA/SirajMPRA_concat.csv", low_memory=False)
# Siraj_df
Siraj_df = pd.read_csv("../../data/SirajMPRA/SirajMPRA_raw.csv", low_memory=False)
Siraj_df.rename(columns={'Allele 1 Oligo': 'seq'}, inplace=True)

Siraj_df[['chr', 'pos', 'ref', 'alt']] = Siraj_df['Variant'].str.split(':', expand=True)
Siraj_df['pos'] = Siraj_df['pos'].astype(int)
Siraj_df
```

```python
merged_df = pd.merge(Gosai_df, Siraj_df, how='inner', on='seq')
merged_df.rename(columns=lambda x: x.replace('HEPG2', 'HepG2'), inplace=True)
merged_df
```

```python
merged_df = merged_df.sort_values(by=['chr_y', 'pos'])
merged_df
```

```python
for cell_type in ['HepG2', 'K562', 'SKNSH']:
    x = merged_df[f'{cell_type}_mean']
    y = merged_df[f'allele1.log2FC_{cell_type}']
    plt.figure(figsize=(4, 4))
    sns.scatterplot(x=x[:1000], y=y[:1000], s=1)
    print(cell_type, 'pearson r =', pearson(x, y))
    
    # for chr in range(1, 23):
    #     x_chr = x[merged_df['chr_y']=='chr'+str(chr)]
    #     y_chr = y[merged_df['chr_y']=='chr'+str(chr)]
    #     print(cell_type, 'chr', chr, 'pearson r =', pearson(x_chr, y_chr))
```
