```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm

import sys
sys.path.append('/home/hxcai/cell_type_specific_CRE')

import utils as utils
from scipy.stats import pearsonr
```

```python
sei_model_dir = '/home/hxcai/cell_type_specific_CRE/sei_framework/model'
clustervfeat = np.load(os.path.join(sei_model_dir, 'projvec_targets.npy'))

sei_joint_embedding = np.load('./data/sei_joint_embedding.npy')
sei_joint_chromatin_pred = np.load('./data/sei_joint_chromatin_pred.npy')
sei_joint_class_pred = np.dot(sei_joint_chromatin_pred, clustervfeat.T) / np.linalg.norm(clustervfeat, axis=1)
print(f'{sei_joint_chromatin_pred.shape=}, {sei_joint_embedding.shape=}, {sei_joint_class_pred.shape=}')
```

```python
chromatin_names = utils.load_txt(os.path.join(sei_model_dir, 'target.names'))
class_names = utils.load_txt(os.path.join(sei_model_dir, 'seqclass.names'))

df = pd.read_csv('../data/lentiMPRA/joint_library_table_s.csv')
# seqs = df['seq'].to_numpy()
H_labels = df['HepG2'].to_numpy()
K_labels = df['K562'].to_numpy()


indices = [i for i, name in enumerate(chromatin_names) if "HepG2" in name]
r_list = []
for i in tqdm(indices):
    r = pearsonr(sei_joint_chromatin_pred[:, i], H_labels)[0]
    r_list.append(r)
sort_index = np.argsort(r_list)[::-1]
sort_value = np.sort(r_list)[::-1]
for i in range(20):
    print(sort_value[i], indices[sort_index[i]], chromatin_names[indices[sort_index[i]]])


indices = [i for i, name in enumerate(chromatin_names) if "K562" in name]
r_list = []
for i in tqdm(indices):
    r = pearsonr(sei_joint_chromatin_pred[:, i], K_labels)[0]
    r_list.append(r)
sort_index = np.argsort(r_list)[::-1]
sort_value = np.sort(r_list)[::-1]
for i in range(20):
    print(sort_value[i], indices[sort_index[i]], chromatin_names[indices[sort_index[i]]])




r_list = []
for i in range(len(class_names)):
    r = pearsonr(sei_joint_class_pred[:, i], H_labels)[0]
    r_list.append(r)
sort_index = np.argsort(r_list)[::-1]
sort_value = np.sort(r_list)[::-1]
for i in range(10):
    print(sort_value[i], sort_index[i], class_names[sort_index[i]])


r_list = []
for i in range(len(class_names)):
    r = pearsonr(sei_joint_class_pred[:, i], K_labels)[0]
    r_list.append(r)
sort_index = np.argsort(r_list)[::-1]
sort_value = np.sort(r_list)[::-1]
for i in range(10):
    print(sort_value[i], sort_index[i], class_names[sort_index[i]])
```
