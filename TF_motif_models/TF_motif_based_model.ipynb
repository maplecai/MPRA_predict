```python
TF_counts_df = TF_counts_df.loc[:, list(TF_RNA_df['gene_symbol'])]
TF_counts_df
```

```python
cell_type = 'HepG2'
data_split = 'train'

selected_index = data_split_index_dict[data_split]
X_train = TF_counts_df[selected_index]
y_train = seq_exp_df[selected_index][cell_type]

model = Lasso(alpha=0.01)
model.fit(X_train, y_train)

for cell_type in ['HepG2', 'K562', 'SKNSH']:
    for data_split in ['train', 'valid', 'test', 'designed']:
        print(cell_type, data_split)
        selected_index = data_split_index_dict[data_split]
        X_valid = TF_counts_df[selected_index]
        y_valid = seq_exp_df[selected_index][cell_type]
        y_pred = model.predict(X_valid)
        r = pearsonr(y_valid, y_pred)[0]
        print(r)
    print()
```
